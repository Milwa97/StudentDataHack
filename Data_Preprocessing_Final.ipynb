# Transform categorical variables int 'dummy' variable (one-hot encoding)

```python
stars['smoking'] = stars['smoking'].astype(object)
stars['drinker'] = stars['drinker'].astype(object)
stars['dress_preference'] = stars['dress_preference'].astype(object)
stars['company'] = stars['company'].astype(object)
stars['transport'] = stars['transport'].astype(object)
stars['marital_status'] = stars['marital_status'].astype(object)
stars['interest'] = stars['interest'].astype(object)
stars['religion'] = stars['religion'].astype(object)
stars['activity'] = stars['activity'].astype(object)
stars['budget'] = stars['budget'].astype(object)
stars['cuisine_x'] = stars['cuisine_x'].astype(object)
stars['cuisine_y'] = stars['cuisine_y'].astype(object)
stars['payment'] = stars['payment'].astype(object)
stars['color'] = stars['color'].astype(object)
stars['parking_lot'] = stars['parking_lot'].astype(object)
stars['payment'] = stars['payment'].astype(object)


# Some of the predictors are categorical. Lets produce some dummy variables
categorical_bolean = (stars.dtypes == "object").values
data_numeric = stars.loc[:, ~categorical_bolean]
data_categorical = stars.loc[:, categorical_bolean]
data_dummies = pd.get_dummies(data_categorical)
data = pd.concat([data_numeric, data_dummies], axis=1)
data
```
